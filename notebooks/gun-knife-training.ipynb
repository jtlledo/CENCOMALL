```python
# Dataset: https://www.universe.roboflow.com/joao-assalim-xmovq/weapon-2/dataset/2 
from ultralytics import YOLO
import os

os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'

model = YOLO("../src/models/yolov8n.pt")
model.to("cuda")
data_yaml_path = os.path.abspath(os.path.join(os.path.dirname("__file__"), "../data/data.yaml"))
print(f"Using data configuration from: {data_yaml_path}")

results = model.train(
    data=data_yaml_path, 
    epochs=100, 
    imgsz=512,
    batch=4,
    cache='disk'
)
```

```python
# Save the model to a file
model.save("yolov8-knife.pt")
```

```python
import cv2

use_webcam = True
CONFIDENCE_THRESHOLD = 0.5

if use_webcam:
    cap = cv2.VideoCapture(0)
    assert cap.isOpened(), "Failed to open webcam"
else:
    video_path = "../tests/knife.mp4"
    cap = cv2.VideoCapture(video_path)
    assert cap.isOpened(), f"Failed to open video file at {video_path}"

if use_webcam:
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

print("Press 'q' to exit")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)
    display_frame = frame.copy()
    
    for r in results:
        boxes = r.boxes
        for box in boxes:
            confidence = box.conf[0].item()
            
            if confidence >= CONFIDENCE_THRESHOLD:
                x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()
                x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
                
                class_id = int(box.cls[0].item())
                class_name = model.names[class_id]
                
                cv2.rectangle(display_frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                label = f"{class_name}: {confidence:.2f}"
                cv2.putText(display_frame, label, (x1, y1 - 10), 
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
                
                print(f"Detected {class_name} with confidence {confidence:.2f} at [{x1}, {y1}, {x2}, {y2}]")
    
    cv2.imshow("YOLO Detection", display_frame)
    
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
