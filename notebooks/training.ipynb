```python
from ultralytics import YOLO
import os

# Load a COCO-pretrained YOLO11n model
model = YOLO("../src/models/yolo11x.pt")
model.to("cuda")
data_yaml_path = os.path.abspath(os.path.join(os.path.dirname("__file__"), "../data/data.yaml"))
print(f"Using data configuration from: {data_yaml_path}")

# Train the model on the COCO8 example dataset for 100 epochs
results = model.train(data=data_yaml_path, epochs=100, imgsz=640)
```
